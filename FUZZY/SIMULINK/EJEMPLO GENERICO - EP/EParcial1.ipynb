```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

```python
#entradas
e = ctrl.Antecedent(np.arange(-100, 100, 0.01), 'e')
de = ctrl.Antecedent(np.arange(-4, 4, 0.01), 'de')
#salidas
P = ctrl.Consequent(np.arange(-3, 3, 0.01),'P', defuzzify_method= 'centroid')
```

```python
# Definir las funciones de membresia para e
e['eNG'] = fuzz.trimf(e.universe, [-100, -75, -50])
e['eNM'] = fuzz.trimf(e.universe, [-75, -50, -25])
e['eNP'] = fuzz.trimf(e.universe, [-50, -25, 0])
e['eZ'] = fuzz.trimf(e.universe, [-25, 0, 25])
e['ePP'] = fuzz.trimf(e.universe, [0, 25, 50])
e['ePM'] = fuzz.trimf(e.universe, [25, 50, 75])
e['ePG'] = fuzz.trimf(e.universe, [50, 75, 100])
```

```python
e.view()
```

```python
# Definir las funciones de membresia para DE
de['deNG'] = fuzz.trimf(de.universe, [-4, -3, -2])
de['deNM'] = fuzz.trimf(de.universe, [-3, -2, -1])
de['deNP'] = fuzz.trimf(de.universe, [-2, -1, 0])
de['deZ'] = fuzz.trimf(de.universe, [-1, 0, 1])
de['dePP'] = fuzz.trimf(de.universe, [0, 1, 2])
de['dePM'] = fuzz.trimf(de.universe, [1, 2, 3])
de['dePG'] = fuzz.trimf(de.universe, [2, 3, 4])
```

```python
de.view()
```

```python
# Definir las funciones de membresia para P
P['UFNG'] = fuzz.trimf(P.universe, [-3, -2.25, -1.5])
P['UFNM'] = fuzz.trimf(P.universe, [-2.25, -1.5, -0.75])
P['UFNP'] = fuzz.trimf(P.universe, [-1.5, -0.75, 0])
P['UFZ'] = fuzz.trimf(P.universe, [-0.75, 0, 0.75])
P['UFPP'] = fuzz.trimf(P.universe, [0, 0.75, 1.5])
P['UFPM'] = fuzz.trimf(P.universe, [0.75, 1.5, 2.25])
P['UFPG'] = fuzz.trimf(P.universe, [1.5, 2.25, 3])
```

```python
P.view()
```

```python
#definimos reglas
rule1 = ctrl.Rule(e['eNG'] & de['deNG'], [P['UFNG']])
rule2 = ctrl.Rule(e['eNG'] & de['deNM'], [P['UFNG']])
rule3 = ctrl.Rule(e['eNG'] & de['deNP'], [P['UFNG']])
rule4 = ctrl.Rule(e['eNG'] & de['deZ'], [P['UFNG']])
rule5 = ctrl.Rule(e['eNG'] & de['dePP'], [P['UFNM']])
rule6 = ctrl.Rule(e['eNG'] & de['dePM'], [P['UFNP']])
rule7 = ctrl.Rule(e['eNG'] & de['dePG'], [P['UFZ']])

#definimos reglas
rule8 = ctrl.Rule(e['eNM'] & de['deNG'], [P['UFNG']])
rule9 = ctrl.Rule(e['eNM'] & de['deNM'], [P['UFNG']])
rule10 = ctrl.Rule(e['eNM'] & de['deNP'], [P['UFNG']])
rule11 = ctrl.Rule(e['eNM'] & de['deZ'], [P['UFNM']])
rule12 = ctrl.Rule(e['eNM'] & de['dePP'], [P['UFNP']])
rule13 = ctrl.Rule(e['eNM'] & de['dePM'], [P['UFZ']])
rule14 = ctrl.Rule(e['eNM'] & de['dePG'], [P['UFPP']])
#definimos reglas
rule15 = ctrl.Rule(e['eNP'] & de['deNG'], [P['UFNG']])
rule16 = ctrl.Rule(e['eNP'] & de['deNM'], [P['UFNG']])
rule17 = ctrl.Rule(e['eNP'] & de['deNP'], [P['UFNM']])
rule18 = ctrl.Rule(e['eNP'] & de['deZ'], [P['UFNP']])
rule19 = ctrl.Rule(e['eNP'] & de['dePP'], [P['UFZ']])
rule20 = ctrl.Rule(e['eNP'] & de['dePM'], [P['UFPP']])
rule21 = ctrl.Rule(e['eNP'] & de['dePG'], [P['UFPM']])
#definimos reglas
rule22 = ctrl.Rule(e['eZ'] & de['deNG'], [P['UFNG']])
rule23 = ctrl.Rule(e['eZ'] & de['deNM'], [P['UFNM']])
rule24 = ctrl.Rule(e['eZ'] & de['deNP'], [P['UFNP']])
rule25 = ctrl.Rule(e['eZ'] & de['deZ'], [P['UFZ']])
rule26 = ctrl.Rule(e['eZ'] & de['dePP'], [P['UFPP']])
rule27 = ctrl.Rule(e['eZ'] & de['dePM'], [P['UFPM']])
rule28 = ctrl.Rule(e['eZ'] & de['dePG'], [P['UFPG']])

#definimos reglas
rule29 = ctrl.Rule(e['ePP'] & de['deNG'], [P['UFNM']])
rule30 = ctrl.Rule(e['ePP'] & de['deNM'], [P['UFNP']])
rule31 = ctrl.Rule(e['ePP'] & de['deNP'], [P['UFZ']])
rule32 = ctrl.Rule(e['ePP'] & de['deZ'], [P['UFPP']])
rule33 = ctrl.Rule(e['ePP'] & de['dePP'], [P['UFPM']])
rule34 = ctrl.Rule(e['ePP'] & de['dePM'], [P['UFPG']])
rule35 = ctrl.Rule(e['ePP'] & de['dePG'], [P['UFPG']])
#definimos reglas
rule36 = ctrl.Rule(e['ePM'] & de['deNG'], [P['UFNP']])
rule37 = ctrl.Rule(e['ePM'] & de['deNM'], [P['UFZ']])
rule38 = ctrl.Rule(e['ePM'] & de['deNP'], [P['UFPP']])
rule39 = ctrl.Rule(e['ePM'] & de['deZ'], [P['UFPM']])
rule40 = ctrl.Rule(e['ePM'] & de['dePP'], [P['UFPM']])
rule41 = ctrl.Rule(e['ePM'] & de['dePM'], [P['UFPG']])
rule42 = ctrl.Rule(e['ePM'] & de['dePG'], [P['UFPG']])
#definimos reglas
rule43 = ctrl.Rule(e['ePG'] & de['deNG'], [P['UFZ']])
rule44 = ctrl.Rule(e['ePG'] & de['deNM'], [P['UFPP']])
rule45 = ctrl.Rule(e['ePG'] & de['deNP'], [P['UFPM']])
rule46 = ctrl.Rule(e['ePG'] & de['deZ'], [P['UFPG']])
rule47 = ctrl.Rule(e['ePG'] & de['dePP'], [P['UFPG']])
rule48 = ctrl.Rule(e['ePG'] & de['dePM'], [P['UFPG']])
rule49 = ctrl.Rule(e['ePG'] & de['dePG'], [P['UFPG']])
```

```python
reglas_gen = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule30,rule31,rule32,rule33,rule34,rule35,rule36,rule37,rule38,rule39,rule40,rule41,rule42,rule43,rule44,rule45,rule46,rule47,rule48,rule49])
reglas = ctrl.ControlSystemSimulation(control_system=reglas_gen, clip_to_bounds=True)
```

```python
reglas.input['e'] = 55
reglas.input['de']= 3

reglas.compute()
print(reglas.output['P'])

P.view(sim=reglas)
```
