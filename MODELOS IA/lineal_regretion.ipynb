```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
datos = pd.read_csv('Solido_Oxigeno.csv')
datos

x  = datos[["Reduccion de solidos"]]
x

y = datos[["Reduccion de la demanda de oxigeno"]]
y

plt.scatter(x, y)
plt.xlabel("Reduccin de solidos")
plt.ylabel("Reduccion de la demanda de oxigeno")
plt.grid()
plt.show()

matriz = datos.to_numpy()
matriz

n = len(matriz)
sumatoria_x = np.sum(matriz[:, 0])
sumatoria_y = np.sum(matriz[:,1])
sumatoria_producto = np.sum(matriz[:, 0] * matriz[:, 1])
sumatoria_cuadrado_x = np.sum(matriz[:,0] * matriz[:, 0])

print("n:", n)
print("sumatoria x:", sumatoria_x)
print("sumatoria y:", sumatoria_y)
print("sumatoria xy:", sumatoria_producto)
print("sumatoria x^2:", sumatoria_cuadrado_x)

b1 = (n*sumatoria_producto - sumatoria_x*sumatoria_y) / (n*sumatoria_cuadrado_x - sumatoria_x*sumatoria_x)
b0 = (sumatoria_y - b1*sumatoria_x) / n

print("b1:", b1)
print("b0:", b0)

clf = LinearRegression()
# Entrenamiento con los datos x, y obtenido del archivo .csv
clf.fit(x,y) 
# Obtener la pendiente del modelo
clf.coef_
# Cruce por el eje de las ordenadas
clf.intercept_

# Probando el mdelo generado con valores especificos
clf.predict([[7]])  
clf.predict([[100]])

# Predecir todos los valores de x con el mdelo generado
predicciones = clf.predict(x)

plt.plot(x, y)
plt.plot(x, predicciones)
plt.title("Regresion Lineal Simple")
plt.xlabel("Reduccion de solidos")
plt.ylabel("Reduccion de la demanda de oxigeno")
plt.legend(["y", "predicciones"])
plt.grid()
plt.show()

plt.scatter(x,y)
plt.plot(x, predicciones)
plt.title("Regresion Lineal Simple")
plt.xlabel("Reduccion de solidos")
plt.ylabel("Reduccion de la demanda de oxigeno")
plt.grid()
plt.show()

mean_squared_error(y, predicciones)
```
